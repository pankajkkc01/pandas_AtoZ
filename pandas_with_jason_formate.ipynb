read json_(read_jason)
lets see how it works

```python
data='{"employee_name":"james","email":"abc@gmail.com","job_profile":[{"title1":"team_lead","title2":"sr_developer"}]}'
```

```python
type(data)#note= json file stored as key value pair something like str
```

to check is this a valid json or note do check it on jsonlint.com

```python
import pandas as pd
```

```python
pd.read_json(data)# now json converted into the dataframe
```

```python
pd.read_json(data,orient="record")#hence it will take record as argument  by defoult
```

```python
pd.read_json(data,orient="index")# this is diffrent here column names converted into the row index
```

now convert dataframe into the json formate

```python
import pandas as pd
df=pd.DataFrame([['a','b'],['c','d']],
                index=['row1','row2'],# this specifies the row index
                columns=['col1','col2'])# this specifies the columns indexes
```

```python
print(df)
```

```python
df# this show the data frame
```

```python
df.to_json()# this convert data frame into json formate
```

```python
df.to_json(orient="index")# this row index as key and pul col and value as values
```

```python
df.to_json(orient="columns")# this will make column index as key and vise versa
```

```python
df.to_json(orient="records")# these are the diffrent ways of storing json file into difffrent json formate
```

```python
df.to_json(orient="split")# it just splits the string into substrings
```

```python
df.to_json(orient="table")# give diifrent information abut the  data
```

```python
schema='{"schema":{"fields":[{"name":"index","type":"string"},{"name":"col1","type":"string"},{"name":"col2","type":"string"}],"primaryKey":["index"],"pandas_version":"1.4.0"},"data":[{"index":"row1","col1":"a","col2":"b"},{"index":"row2","col1":"c","col2":"d"}]}'
```

```python
pd.read_json(schema,orient="table")# converting back schema into the dataframe
```
